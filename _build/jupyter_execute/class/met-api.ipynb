```python
# 2. "OBJECT" ENDPOINT
object_path = '/public/collection/v1/objects/'
new_first_object = requests.get(root + object_path + str(new_first))
```

```python
ids = new_parsed['objectIDs']
```

```python
# write a loop that makes an "object" request for the first 
# 10 items in the "search" results
first_ten = []
for i in ids[:100]:
    # make a new request 
    response = requests.get(root + object_path + str(i))
    parsed = response.json()
    first_ten.append(parsed)
```
